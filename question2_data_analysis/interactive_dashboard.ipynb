```python
import os
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

CLEANED_DATA_PATH = "./data/cleaned/"
file_path = os.path.join(CLEANED_DATA_PATH, "cleaned_books_data_500.csv")
df = pd.read_csv(file_path)

categories = sorted(df["category"].unique())
cat_dropdown = widgets.Dropdown(options=["All"] + categories, description="Category:")
min_price = widgets.FloatSlider(value=float(df["price"].min()), min=float(df["price"].min()),
                                max=float(df["price"].max()), step=1.0, description="Min £")
max_price = widgets.FloatSlider(value=float(df["price"].max()), min=float(df["price"].min()),
                                max=float(df["price"].max()), step=1.0, description="Max £")

out = widgets.Output()

def update(*args):
    with out:
        out.clear_output()
        d = df.copy()
        if cat_dropdown.value != "All":
            d = d[d["category"] == cat_dropdown.value]
        d = d[(d["price"] >= min_price.value) & (d["price"] <= max_price.value)]

        fig = px.histogram(d, x="price", nbins=20, title="Interactive Price Distribution")
        
        fig.update_layout(
            xaxis_title="Price",
            yaxis_title="Count"
        )
        
        fig.show()

for w in [cat_dropdown, min_price, max_price]:
    w.observe(update, "value")

display(widgets.VBox([cat_dropdown, min_price, max_price, out]))
update()
```
