```python
import os
import pandas as pd
import plotly.express as px

CLEANED_DATA_PATH = "./data/cleaned/"
file_path = os.path.join(CLEANED_DATA_PATH, "cleaned_books_data_500.csv")
df = pd.read_csv(file_path)

fig = px.scatter(
    df, x="rating", y="price", color="category",
    hover_data=["title", "availability", "price_category"],
    title="Interactive Price vs Rating (Hover + Category Colors)",
    trendline="ols"
)
fig.show()
```
